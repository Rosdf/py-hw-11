```python
#задание 2
data = pd.read_csv('power.csv')
```

```python
countries = ['Latvia', 'Lithuania', 'Estonia']
dates = range(2005, 2011)
categories = [4, 12, 21]
balt_data = data[(data.country.isin(countries)) & (data.year.isin(dates)) & (data.category.isin(categories)) & (data.quantity >= 0)]
```

```python
balt_data.groupby(['country']).sum()['quantity']
```
